```python
# import random package
import random

# Description of the game situation
print(f"""
In this scenario, you are the CEO of Company X and you will 
have to carefully consider your options and weigh the potential benefits and 
drawbacks of each decision. There are two companies, X and Y, that are selling 
similar products in the same market. Both companies have the option to either 
lower their prices or keep them the same. Company X and Company Y will have to decide 
whether to lower their prices in an effort to gain an advantage over the other, or to 
keep their prices the same in order to maintain a moderate profit.
          
Company X may decide to lower their prices in an effort to gain market share and increase 
their profits. However, if Company Y also lowers their prices, both companies will face 
increased competition and will make less profit. This will put both companies in a difficult 
situation, as they will have to balance their desire for increased profits with the reality 
of increased competition.""")

# Posible answers from the user
pos_answer = ["yes","y","ok","sure","Yes","Ok"]
neg_answer = ["no","n","nay","nah","No","n"]

validation = 0

# Loop until user enters a valid answer for playing the game
while validation == 0:
    # Ask if the user wants to play game theory
    user_play = input(f"""Do you want to play a Game Theory Scenario? """).lower()
    if user_play in pos_answer:
        validation += 1
        user_play = "yes"
    elif user_play in neg_answer:
        validation += 1
        user_play = "no"
    else:
        print('''
        Try another answer silly!
        ''')
        
# Create a board game with the pay-off matrix.
board_game = [[4, 10], [5, 6]]
    
# Save board points and results in a dictionary. R: remain silent, S: betray
dictionary_results = {"(L, L)": (board_game[0][0], board_game[0][0]),
                      "(L, H)": (board_game[0][1], board_game[1][0]),
                      "(H, L)": (board_game[1][0], board_game[0][1]),
                      "(H, H)": (board_game[1][1], board_game[1][1])}

# If user want to play
if "yes" in user_play:
    user_name = input("Hi, and welcome to this game! You will play against me! Whats your name? : ").capitalize()
    
    print("""
    You will have to decide if your prices are High or Low in order to achieve 
    your companies profit goals for the year.""")
    
    # Board Game
    print(f"""
    CFO: {user_name}, this is the situation, think well before deciding.
    \n
    Company X \ Company Y\t\t Low Price \t\t\t  High Price
    --------------\t --------------------------\t ---------------------------
    Low Price\t\t Both get $4MM\t\t            Y gets $5MM, X $10MM
    High Price \t\t Y gets $10MM, X $5MM\t            Both get $6MM
    """)
    

    # prompt the user to select what to do
    user_choose = input("""
    What would you do? Would you set prices high or low?
    Select:
    - 'L' to set Low Prices
    - 'H' to set High Prices
    """).upper()
    
    # Computer (Python) chosee randomly either to remaint silent or betray
    computer_choose = random.randint(0,1) # 0: Low Prices; 1: High Prices
    
    # conditional structure with the outputs of each scenario
    if user_choose == "L" and computer_choose == 0:
        user_result = dictionary_results["(L, L)"][0]
        computer_result = dictionary_results["(L, L)"][1]
        print(f"""
    You've decided to set the prices Low, as well as Company Y!
    Company X earned ${user_result}MM in profit.
    Company Y earned ${computer_result}MM in profit.""")
    
    elif user_choose == "H" and computer_choose == 0:
        user_result = dictionary_results["(H, L)"][0]
        computer_result = dictionary_results["(H, L)"][1]
        print(f"""
    You've decided to set the prices High!, Company Y will set them Low!
    Company X earned ${user_result}MM in profit.
    Company Y earned ${computer_result}MM in profit""")

    elif user_choose == "L" and computer_choose == 1:
        user_result = dictionary_results["(L, H)"][0]
        computer_result = dictionary_results["(L, H)"][1]
        print(f"""
    You've decided to set the prices Low!, Company Y will set them High!
    Company X earned ${user_result}MM in profit.
    Company Y earned ${computer_result}MM in profit""")

    elif user_choose == "S" and computer_choose == 1:
        user_result = dictionary_results["(H, H)"][0]
        computer_result = dictionary_results["(H, H)"][1]
        print(f"""
    You've decided to set the prices High, as well as Company Y!
    Company X earned ${user_result}MM in profit.
    Company Y earned ${computer_result}MM in profit.""")

    # Declaring a winner.

    if user_result > computer_result:
        print(f"""Congrats, Company X made more profits than Company Y!""")

    elif user_result == computer_result:
        print(f"""Both companies made the same profit! Both Winners""")

    elif user_result <= computer_result:
        print(f"""Congrats to Company Y, You are Fired!""")
        
elif "no" in user_play:
    print("You don't want to play as the CEO, so let the computer make the decisiones for you")
    
    rounds = ("Round 1", "Round 2", "Round 3", "Round 4", "Round 5")
    c1_score=0
    c2_score=0
    
    for i in rounds:
        
        computer_1 = random.randint(0,1) # 0: Low Price; 1: High Price
        
        import datetime
        current_time = datetime.datetime.now().time()
        noon = datetime.time(12)

        if current_time < noon:
            computer_2 = 1
        else:
            computer_2 = 0

         # conditional structure with the outputs of each scenario
        if computer_1 == 0 and computer_2 == 0:
            c1_result = dictionary_results["(L, L)"][0]
            c2_result = dictionary_results["(L, L)"][1]
            c1_score = c1_score + c1_result
            c2_score = c2_score + c2_result
            print(f"""{i} \t
        Company X set the prices Low, as well as Company Y!
        Company X earned ${c1_result}MM in profit.
        Company Y earned ${c2_result}MM in profit.""")
    
        elif computer_1 == 1 and computer_2 == 0:
            c1_result = dictionary_results["(H, L)"][0]
            c2_result = dictionary_results["(H, L)"][1]
            c1_score = c1_score + c1_result
            c2_score = c2_score + c2_result
            print(f"""{i} \t
        Company X set the prices High, Company Y kept them Low!
        Company X earned ${c1_result}MM in profit.
        Company Y earned ${c2_result}MM in profit.""")

        elif computer_1 == 0 and computer_2 == 1:
            c1_result = dictionary_results["(L, H)"][0]
            c2_result = dictionary_results["(L, H)"][1]
            c1_score = c1_score + c1_result
            c2_score = c2_score + c2_result
            print(f"""{i} \t
        Company X set the prices Low, Company Y kept them High!
        Company X earned ${c1_result}MM in profit.
        Company Y earned ${c2_result}MM in profit.""")
              
        elif computer_1 == 1 and computer_2 == 1:
            c1_result = dictionary_results["(H, H)"][0]
            c2_result = dictionary_results["(H, H)"][1]
            c1_score = c1_score + c1_result
            c2_score = c2_score + c2_result
            print(f"""{i} \t
        Company X set the prices High, Company Y kept them High too!
        Company X earned ${c1_result}MM in profit.
        Company Y earned ${c2_result}MM in profit.""")
    
    print(f"""
    At the end of all the five rounds, 
    - Company X managed to earn a total of ${c1_score}MM
    - Company Y managed to earn a total of ${c2_score}MM""")

    # Declaring a winner.

    if c1_score > c2_score:
        print(f"""Congrats, Company X made more profits than Company Y!""")

    elif c1_score == c2_score:
        print(f"""Both companies made the same profit! Both winners""")

    elif c1_score <= c2_score:
        print(f"""Congrats to Company Y, You are Fired!""")
    

    
```
