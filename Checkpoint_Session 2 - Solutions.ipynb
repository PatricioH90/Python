Our database appears to be in a wide data format; for a better analysis, let's convert it to a long format. <br>
Name your variable 'origin,' and your values 'points,' respectively. <br>
Only the first 30 wineries will be displayed.

```python
wine_long = wine_df.melt(
    id_vars=['id','winery', 'variety', 'price', 'bottles'],
    var_name='origin',
    value_name='points')
wine_long
```

How big is the wine database now? <br>

```python
wine_long.shape
```

<br>
Excellent work!! We now have a large database, but because the feature "origin" still links country and province, let's split it into two columns: 'country' and 'province.'

```python
wine_long[['country', 'province']] = wine_long['origin'].str.split('_', expand=True)
wine_long
```

<br> We don't really need the Origin feature with this, so let's drop it.

```python
wine_long.drop(columns=['origin'], inplace=True)

wine_long
```

<br> Calculate the total amount

```python
wine_long['total'] = wine_long['price']*wine_long['bottles']
wine_long
```

<br> Finally, let's pivot our database by country

```python
wine_tidy = wine_long.pivot_table(
    index=['id','winery','variety','price','bottles','total','province'],
    columns='country',
    values='points')
wine_tidy
```

How many points does Chile has?

```python
wine_tidy.loc[:,('chile'),].dropna()
```

<br> Our distribution center, is asking for the total number of bottles that we have in Italy.

```python
italy_wine = wine_tidy.loc[wine_tidy['italy'].notnull()].reset_index()
italy_wine['bottles'].sum()
```
