## Problem 1

```python
#Creating a List of the Target Countries to include in the Filtered DataFrame
target_coun = ["USA","CAN","AUS"]

#Creating a new DataFrame that only include data from US, Canada and Australia
unemploy_filter = unemploy_path[unemploy_path["LOCATION"].isin(target_coun)]

#Create a new DataFrame that will Aggregate the Functions of Min, Max and Average for the Unemployment Index
unemploy_table = unemploy_filter.groupby("LOCATION")["Value"].agg([np.min, np.max, np.mean]).round(2)

# Create a bar plot of the Index Values By Country
unemploy_table.plot(kind="bar")
plt.title("Unemployment Index - Min, Max and Average - By Country")

# Show Table and plot
print(unemploy_table)

print("""
-------------------------------------------------------------------------------------------------------------
""")

plt.show()

print("""
- Australia has the lowest minimum value of 4.92, while the United States has the lowest maximum value of 3.40.
- Canada has the highest maximum value of 14.00 and the highest mean value of 9.70.
- Overall, the US Average Unemployment is stable""")
```

## Problem 2

```python
#Create a New Column in the DataFrame
pp_path["Extra Purchase Power"] = pp_path["Local Purchasing Power Index"] - pp_path["Cost of Living Plus Rent Index"]

#Create a New DataFrame with the Columns needed for the Analysis
pp_table = pp_path[["Country","Cost of Living Plus Rent Index","Local Purchasing Power Index","Extra Purchase Power"]]

#List of Targeted Countries
target_countries = ["United States","Canada","Australia"]

#Create a New DataFrame that only includes the Targeted Countries
pp_final= pp_table[pp_table["Country"].isin(target_countries)]

# Get the 3 Index Values By Country
pp_graph = pp_final.groupby("Country").sum()

# Create a bar plot of the Index Values By Country
pp_graph.plot(kind="bar")
plt.title("Economic Capability Index by Country")

print("""-------------------------------------------------------------------------------
This graph compared by country the 3 Indexes mentioned in the Legend.
On average the US has a higher Purchase Power and Cost of Living Indexes
-------------------------------------------------------------------------------""")
# Show the plot
plt.show()

#Get the Extra Purchase Power Index to Compare
pp_ep_graph = pp_final.groupby("Country")["Extra Purchase Power"].sum()

# Create a bar plot of the Index Values By Country
plt.ylim(0, 100)
pp_ep_graph.plot(kind="bar")
plt.title("Extra Purchase Power Index by Country")

print("""-------------------------------------------------------------------------------
This graph is a zoom in of the Extra Purchase Power Index by Country
-------------------------------------------------------------------------------""")
# Show the plot
plt.show()
```
