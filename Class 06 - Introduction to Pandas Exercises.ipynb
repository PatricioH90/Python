<h4> 1.3. List of files</h4>
You can access each element of a directory by using the list directory method from the os package.

```python
# Code 1.3.1.

# Import listdir method
from os import listdir

# printing all files present in the working directory
for item in listdir(path = "."): # . means current directory
    print(item)
```

```python
# Code 1.3.2.
# Moving one folder deeper
for item in listdir(path = "./__resources"): 
    print(item)
```

```python
# Code 1.3.3.
# Moving one folder before (parent)
for item in listdir(path = ".."): # two dots == parent directory
    print(item)
```

<h3>6. Adding new columns</h3><br>
Always keep the original DataFrame values. Therefore, create a new DataFrame using the .copy() method

~~~
df_copy = df.copy()
~~~

```python
# Code 6.1. Bad Practice!
# Declare a new DataFrame
df = pd.DataFrame([[1,"a"], [2, "b"], [3, "c"]], columns = ["numbers", "letters"])

print(f"""DataFrame before 'copy': 
{df}""")

df_copy = df

df_copy["special_characters"] = [".", "/", "*"]

print(f"""\n
- Original DataFrame: 
{df}
- Copy DataFrame: 
{df_copy}""")
```
