```python
#Author: Alvaro Ronquillo

#Import Package for Winning Reward
import webbrowser
import turtle

#Defining URL Links (Related to the Ending of the Game)
url = "https://www.youtube.com/watch?v=dQw4w9WgXcQ&ab_channel=RickAstley"
url_2 = "https://www.youtube.com/watch?v=R52UhAM8jZg&ab_channel=NosideFoxhound"
url_3 = "https://www.youtube.com/watch?v=-ZGlaAxB7nI&ab_channel=iwanPlays"

#Asking the User Name
user_name = input("Tell us your name brave hero?  ").title()

#Introduction to the Story
print("""\nThe story of Zelda is one of adventure and mystery, set in the land of Hyrule. 
The land is ruled by the benevolent Princess Zelda, but it is also threatened by the evil Ganon, 
who seeks to obtain the powerful Triforce and use its power to conquer the land.
""")

print(f"""{user_name}, is a young hero chosen by the gods to protect the land of Hyrule and defeat Ganon.

{user_name} sets out on a journey to collect the three Spiritual Stones, 
which are said to unlock the entrance to Ganon's Castle and rescue Princess Zelda.

{user_name}, you must find the three Spiritual Stones and defeat Ganon once and for all!!!

Quick! The first Stone can be found in the Kokori Forest!
""")

# Posible answers from the user if continue playing
pos_answer = ["yes","y","ok","sure","Yes","Ok","Y","Sure"]
neg_answer = ["no","n","nay","nah","No","N"]

#Defining Dead Scenario
def die():
    print("""You die!
    
Better Luck Next Time
    """)
    webbrowser.open(url_3)
    again = input("Play again? Yes or No ")
    # Conditional for the Right Answer from User
    if again in pos_answer:
        level_1()
    elif again in neg_answer:
        "You were defeated! Sucker!"
    else:
        print('You were defeated in Life')

#Defining Fail Scenario 
def fail():
    print("""You have failed on rescuing Princess Zelda
    
Better Luck Next Time
    """)
    webbrowser.open(url_2)
    again = input("Play again? Yes or No")
    # Conditional for the Right Answer from User
    if again in pos_answer:
        level_1()
    elif again in neg_answer:
        "You were defeated! Sucker!"
    else:
        print('You were defeated in Life')

#Defining Win Scenario 
def win():
    print(f"""Congratulations {user_name}!!.
You have defeted Ganon and rescued Pricess Zelda!
You win a cookie
    """)
    webbrowser.open(url)

    
#Defining the Game - Starts Here
#Defining Level 1
def level_1():
    print("You have reached the Kokori Forest.!\n")
    input('<Press any key to continue>\n')
    
    print("""You found an allied called the Great Deku Tree.
He explains that the Stone is been guarded by deadly spider
Only a great hero can defeat her and retrieve the Stone.
          
But be aware! Regular weapons aren't very effective against this foe.
          
The Great Deku Tree offers you to choose between three weapons to defeat it
          
A. Bow + Arrows
B. Heavy Armor + Mace
C. Sword + Shield
    """)
   
    choice = input("> ").upper()
    
    if choice == 'A':
        hp = 1
        #Loop until you land the deadly blow
        while hp == 1:
            print("You were able to strike serious damaging blows to the spider foe and left it unconscious.\n")
        
            print(f"""{user_name}! Quick! You only have a few seconds to land a deadly blow! Choose correctly

A. Strike the Chest
B. Strike the Head
C. Strike the Legs
D. Strike the Tail
        
""")
            choice = input("> ").upper()
        
            # Start of nested conditional
            if choice == 'A':
                print(f"""{user_name}, your sword and arrows couldn't pass through the spider's thick skin!
You tried to hide but the foe recovers. You must try again! Keep Fighting!
                """)
                hp = 1
                
            elif choice == 'C':
                print(f"""{user_name}, your sword and arrows couldn't pass through the spider's thick skin!
You tried to hide but the foe recovers. You must try again! Keep Fighting!
                """)
                hp = 1
                
            elif choice == 'D':
                print(f"""{user_name}, your sword and arrows couldn't pass through the spider's thick skin!
You tried to hide but the foe recovers. You must try again! Keep Fighting!
                """)
                hp = 1
            
            elif choice =='B':
                print("""Great Strike! You have defeated the spider and recover the first stone!
Off to the next stone location! The Waterlands, also known as Zora's Domain""")
                hp = 0
                input('<Press any key to continue>\n')
                
                level_2()
            else:
                print("Invalid answer. Try Again")
                hp = 1
            # End of nested conditional
            
    elif choice == 'B':
        print("The armor is to heavy and doesn't let you move to evade the spider's attacks.")   
        die()
        
    elif choice == 'C':
        print("You can not get closer to land a strike with the sword and the spider is to quick for your swings.")        
        die()
    
    else:
        print("Invalid entry. Please try again.\n")
        input('<Press any key to continue>\n')
        
        level_1()
            
#Defining Level 2
def level_2():
    print("You have reached Zora's Domain.!\n")
    input('<Press any key to continue>\n')
    
    print("""You find Pricess Zora and she wishes to help you. 
She explains the Stone is kept inside a huge whale that lives in the nearby seas
The creature is not aggresive but acts defensively if threaten.
          
Finally, you found the whale and felt dazzled by the size of the beast.
          
You must choose how to recover the Stone!
          
A. Kill the whale and open it to retreive the Stone
B. Manage to enter inside the Whale and recover the stone
    """)
   
    
    choice = input("> ").upper()
   
    if choice == 'A':
        
        print("""You were able to strike serious damaging blows to the whale but it wasnt enough to kill it.
The whale flees far away and you loose the chance of recovering the Stone.\n
        """)
        fail()
             
    elif choice == 'B':
        
        print("""You were able to enter inside the whale and recover 
the Water Stone and the whale thanks you for it by rewarding you the location of the final Stone

Off to Fire Mountain for the Last Stone!
        """)
        input('<Press any key to continue>\n')
        
        level_3()  
   
    else:
        
        print("Invalid entry. Please try again.\n")
        input('<Press any key to continue>\n')
        
        level_2()

#Defining Level 3
def level_3():
    print("You have reached Fire Mountain.!\n")
    input('<Press any key to continue>\n')
    
    print("""On your way to the mountain you found that the Gorons, creatures from
area, were waiting for you.

You met with the Goron King to explain the situation and he wishes to help as well. 

The Goron King tells you he is going to give you the stone and also asks if you could
help by bringing some food from the valley.

You must choose
          
A. Grab the Stone and Bring them Food **May reveal final boss secrets*
B. Grab the Stone and leave in a hurry
    """)
   
    choice = input("> ").upper()
   
    if choice == 'A':
        print("""You were able to bring all the food that the Goron King requested.
As a reward, the Goron King reveals that Ganon weakess points are his hands.

You grab the Stone and head to Ganon's Castle
        """)
        input('<Press any key to continue>\n')
        
        level_final()
             
    elif choice == 'B':
        print("You grab the Stone and head to Ganon's Castle in a hurry")
        input('<Press any key to continue>\n')
        
        level_final()  
   
    else:
        print("Invalid entry. Please try again.\n")
        input('<Press any key to continue>\n')
        
        level_3()

def level_final():
    print("You have reached the Ganon's Castle.!\n")
    input('<Press any key to continue>\n')
    
    print(f"""It's too late to stop Ganon, who has already obtained full power
and used it to escape into the Dark World. {user_name} follows Ganon into the Dark World, 
fighting his way through the twisted and dangerous landscape.

{user_name} eventually reaches Ganon's final room, where he must defeat the 
evil lord in a final battle.

You must choose your weapon to fight Ganon
          
A. Master Sword + Hyrule Shield
B. Longsword 
C. Bow + Arrows

    """)
   
    choice = input("> ").upper()
    
    if choice == 'A':
        #Defining value for HP
        hp = 0
        #Loop until you land the deadly blow 
        while hp < 100:
            
            print("You were able to use the power of the Master Sword to left Ganon unconscious.\n")
        
            print(f"""{user_name}! Quick! You only have a few seconds to land a deadly blow! Choose correctly

A. Strike the Chest
B. Strike the Head
C. Strike the Back
D. Strike the Hands

        
""")
            choice = input("> ").upper()
        
            # Start of nested conditional
            if choice == 'A':
                print(f"""{user_name}, your Master Sword was able to damage Ganon but he recovers!
You must try again! Keep Fighting!
                """)
                hp += 50
                
                print(f"""You striked Ganon for 50 HP
Total Damage Inflicted: {hp}%""")
                
            elif choice == 'B':
                print(f"""{user_name}, your Master Sword was able to land a heavy blow on Ganon but he recovers!
You must try again! Keep Fighting!
                """)
                hp += 60
                
                print(f"""You striked Ganon for 60 HP
Total Damage Inflicted: {hp}%""")
                
            elif choice == 'C':
                print(f"""{user_name}, your Master Sword was able to land a minor blow on Ganon and he quickly recovers!
You must try again! Keep Fighting!
                """)
                hp += 30
                
                print(f"""You striked Ganon for 30 HP
Total Damage Inflicted: {hp}%""")
            
            elif choice =='D':
                print(f"""{user_name}, your Master Sword was able to land a deadly blow on Ganon and he is hurt!
Keep Fighting!
                """)
                hp += 70
                
                print(f"""You striked Ganon for 70 HP
Total Damage Inflicted: {hp}%""")
            
            else:
                print("Invalid Answer. Try again")
                hp += 0
            # End of nested conditional
        input('<Press any key to continue>\n')
        win()
        
    elif choice == 'B':
        print("You don't have any defensive armor. Ganon kills you easily.")   
        die()
        
    elif choice == 'C':
        print("Arrows arent strong enough to wound Ganon.")        
        die()
    
    else:
        print("Invalid entry. Please try again.\n")
        input('<Press any key to continue>\n')
        
        level_final()

#Calling the Game Function to Start
level_1()
```
