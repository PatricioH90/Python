#### Database connections

Pandas can directly interact with SQLite databases. For interacting with other database flavors the SQLAlchemy package needs to be installed and used.

Creating a sample database from the vehicles.csv file using the `to_sql()` method. `'test.db'` is the name and path of the database file we want to connect to.

```python
import sqlite3

with sqlite3.connect('test.db') as connection:
    pd.read_csv('vehicles.csv.zip', low_memory=False).to_sql(
        name = 'vehicles',
        con = connection,
        index = False,
        if_exists = 'replace')
```

Querying the database is just as simple using the `read_sql()` method.

```python
with sqlite3.connect('test.db') as connection:
    df2 = pd.read_sql(sql = 'SELECT * FROM vehicles',
                      con = connection)
df2.head()
```

# Tidying data

Hadley Wickham introduced the concept of tidy data. This is a framework to structure data sets so they can be easily analysed and visualized. The characteristica of tidy data are:

- Each variable has its own column
- Each observation has its own row
- Each value has its own cell



## Columns containing values

```python
pew = pd.read_csv('pew.csv')
```

```python
pew
```

This is a shape of data commonly used to look at data.

It is known as pivot-table or also as wide data format.

In order to make this data usable it needs to be changed into long format.

This means having religion, income and count variables in separate columns.

Pandas has the `melt()` method for this. It has four important parameters:

- `id_vars` a list of the columns that shall stay the same
- `value_vars` identifies the columns to melt down - by default all except id_vars
- `var_name` new column name for the melted down column labels - by default this is 'variable'
- `value_name` new column name for the melted down column values - by default this is 'value'

```python
pew_long = pew.melt(id_vars='religion', var_name='income', value_name='count')
pew_long.head()
```

In many cases there won't be a single variable to keep.

```python
bb = pd.read_csv('billboard.csv')
bb.shape
```

```python
bb.head()
```

```python
bb_long = bb.melt(
    id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
    var_name='week',
    value_name='rating')
bb_long.head()
```

## Variables in both rows and columns

```python
weather = pd.read_csv('weather.csv')
weather.head()
```

The data contains minimum and maximum temperatures for each day of the month.

The first step is to melt the Dataframe.

```python
weather_melt = weather.melt(
    id_vars=['id', 'year', 'month', 'element'],
    var_name='day',
    value_name='temp')
weather_melt.head()
```

Now the values of the element column need to be pivoted up.

```python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp'
)
weather_tidy.head()
```

The new Dataframe has a hirarchical index structure. It is easier to work with after this is flattened.

```python
weather_td = weather_tidy.reset_index()
weather_td.head()
```

To do the whole in one step is cleaner and ccan be done through method chaining.

```python
weather_td = (
    weather.melt(
        id_vars=['id', 'year', 'month', 'element'],
        var_name='day',
        value_name='temp')
    .pivot_table(
        index=['id', 'year', 'month', 'day'],
        columns='element',
        values='temp')
    .reset_index()
)
weather_td.head()
```

## Concatanation

Concatination means appending rows or columns to a dataset.

```python
df1 = pd.DataFrame({
    'A': ['a0', 'a1', 'a2', 'a3'],
    'B': ['b0', 'b1', 'b2', 'b3'],
    'C': ['c0', 'c1', 'c2', 'c3'],
    'D': ['d0', 'd1', 'd2', 'd3']})

df2 = pd.DataFrame({
    'A': ['a4', 'a5', 'a6', 'a7'],
    'B': ['b4', 'b5', 'b6', 'b7'],
    'C': ['c4', 'c5', 'c6', 'c7'],
    'D': ['d4', 'd5', 'd6', 'd7']})

df3 = pd.DataFrame({
    'A': ['a8', 'a9', 'a10', 'a11'],
    'B': ['b8', 'b9', 'b10', 'b11'],
    'C': ['c8', 'c9', 'c10', 'c11'],
    'D': ['d8', 'd9', 'd10', 'd11']})
```

Adding rows

```python
row_concat = pd.concat([df1, df2, df3])
row_concat
```

The concat function can also reset the index at the same time.

```python
row_concat = pd.concat([df1, df2, df3], ignore_index=True)
row_concat
```

You can also concatenate columns.

```python
col_concat = pd.concat([df1, df2, df3], axis='columns')
col_concat
```

Ignore index can also be used on column labels.

```python
col_concat = pd.concat([df1, df2, df3], axis='columns', ignore_index=True)
col_concat
```

### Merging the files using a loop

In order to load all files a loop can be used.

```python
from pathlib import Path
bb_files = (
    Path(".")
    .glob("bb/billboard*.csv")
)

# create an empty list to hold the dataframes
bb_list_df = []

# loop through all files
for fn in bb_files:
    bb_list_df.append(pd.read_csv(fn))

# concatenate dataframes
bb_full = pd.concat(bb_list_df)

bb_full.shape
```

### Using list comprehension

Another cleaner option to do the same is using list comprehension:

```python
from pathlib import Path
bb_files = (
    Path(".")
    .glob("bb/billboard*.csv")
)

bb_full = pd.concat([pd.read_csv(data) for data in bb_files])

bb_full.shape
```

## Transformation

Unlike aggregations which take multiple values and return one single vlaue transformations take multiple values and return the same number of values applying a one-to-one transformation.

### Example: z-score

The z-score gives the number of standard deviations a datapoint is away from the mean of the data.

It is calculated as:
$$ z = {x-\mu \over \sigma} $$

with:
- $x$ being the datapoint
- $\mu$ being the average of the dataset
- $\sigma$ being the standard deviation

```python
def zscore(x):
    ''' Calculates the z-Score
    'x' is a vector or series of numeric values
    '''
    return((x - x.mean()) / x.std())
```

To calculate the z-score by year for each life expectancy value in the dataset `groupby()` can be used with `transform()`.

```python
gm['lifeExp_zScore_byYear'] = gm.groupby('year')['lifeExp'].transform(zscore)
gm.head()
```

## Filter

Another operation often performed with grouped data is filtering.

Example:

In the life expectancy dataset you want to only leave data for continents for which we have more than 5 countries worth of data.

```python
gm['country'].value_counts()
```

There are 12 rows per country.

```python
gm['continent'].value_counts()/12
```

Oceania seems to have only 2 countries in this dataset.

```python
gm_filtered = (
    gm
    .groupby('continent')
    .filter(lambda x: x['continent'].count()/12 > 5.0)
)
gm_filtered['continent'].value_counts()/12
```
